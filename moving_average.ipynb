```python
import pandas as pd
from datetime import datetime
import yfinance as yf
data=yf.download('TSLA')
```

```python
data
```

```python
#simple moving average
```

```python
data['Open'].plot(figsize=(15,6))
```

```python
data['Open:10 days rolling']=data['Open'].rolling(window=10,min_periods=1).mean()
```

```python
data[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
```

```python
data['Open:30 days rolling']=data['Open'].rolling(window=30,min_periods=1).mean()
data['Open:50 days rolling']=data['Open'].rolling(window=50,min_periods=1).mean()
```

```python
data[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
data[['Open','Open:50 days rolling']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
data[['Open','Open:30 days rolling']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
```

```python
## expanding 
## CMA
data['Open'].expanding().mean().plot(figsize=(15,6))
```

```python
# EWMA exponential moving average
```

```python
## EMA tesla shares
## Lets smoothing factor -0.1
data['EMA_0.1']=data['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
data[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
```

```python
data['EMA_0.3']=data['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
data[['Open','EMA_0.3']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
```

```python
## EWMA
data['EMA_5days']=data['Open'].ewm(span=5).mean()
```

```python
data[['Open','EMA_0.1','EMA_5days']].plot(xlim=['2020-01-01','2022-02-02'],figsize=(15,6))
```
