```python
 data.resample(rule='A').min()
```

```python
 data.resample(rule='A').max()
```

```python
type(data.resample(rule='A').max())
```

```python
 ##year end frequency
data.resample(rule='A').max()['Open'].plot()
```

```python
##quaterly start frequency
data.resample(rule='B').max()['High'].plot()
```

```python
##Business end frequency
data.resample(rule='BA').max()
```

```python
data.resample(rule='BQS')['Open'].mean().plot(kind='bar')
```

```python
data.resample(rule='M')['Open'].max().plot(kind='bar',figsize=(15,9))
```
